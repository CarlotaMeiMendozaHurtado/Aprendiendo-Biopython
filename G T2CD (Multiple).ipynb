# Ali Multiple

Función objetivo: suma de pares, sumando todas las puntuaciones de los pares posibles usando matriz y penalizando gaps
Algoritmos de ali multiple
1. Programación dinámica
2. Algoritmos heuristicos

1. PD: exacto, usa `Neddleman`
2. AH: inexacto
    - a. `Pregrersivos` [`Clustal`]: ali dos cercnas y añade, si error lo arrastra,usa arbol guia (UPGMA ultrametrico o NEighbout no ultrametrico) 
    - b. `Iterativos` [`Muscle`]: de un progrsivo, refina con los gaps
    - c. `Hibrido` [`Tcoffee`]: combina local y global


- `*` Coin exacta
- `:` Coin alta
- `.` Coin baja

# Ali Mul BioPython

Diferentes formatos de archivo para almacenar secuencias

## Fasta (fas)
Guarda cada secuencia alineada de forma independiente y usa - para gaps

## Stockholm (sto, stk , seed)
formato para la base de datos `PFAM`. Peromite guardar caracteristicas de Ali Mul, `marcas`
- Encabezado obligatorio
- Cierre obligatorio
- Marcas o anotaciones
    - `#=GF` archivo completo
    - `#=GS` anotacion secuencia
    - `#=GR` anotacion residuo
    - `#=GC` anotacion columna

## Clustal (aln, asn)
Secuencia con lineas inferiores de signos
- `*` Coin exacta
- `:` Coin alta
- `.` Coin baja

## Phylip (phy, ph)
Misma longitud
- Linea 1: [num seq][long]
Num max nombe 10 char
Puede ser:
1. interleaved (bloques)
2. sequential ()

# Objeto `MultipleSeqAlignment`
Es como una coleccion donde todo mide los mismo

```python
from Bio.Align import MultipleSeqAlignment
from Bio.SeqRecord import SeqRecord
from Bio.Seq import Seq

alineamiento = MultipleSeqAlignment([
    SeqRecord(Seq("ACTG-A") , id="Seq1"),
    SeqRecord(Seq("ACTGCA") , id="Seq2")
])
print(alineamiento)

print("---- Rebanado")
sub_ali = alineamiento[0:2 , 0:4] # filas columnas el orden fila y el orden col, formas de recorrer cada una
print(sub_ali)

nuevo = alineamiento[: , 0:2] + alineamiento[:, 4:6]
print(alineamiento.substitutions)
```
