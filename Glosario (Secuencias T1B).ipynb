# Básico

````python
from Bio.Seq import Seq
# -------------------------

one = Seq("ATCGAT")

for i,j in enumerate(one,1):
    print(f"{i}{j}")

# ------------------------- 
print("------ Complement ------")
# ------------------------- 

cmp_one = one.complement()
print(cmp_one)

rvcmp_one = one.reverse_complement()
print(rvcmp_one)

# ------------------------- 
print("------ Rebanado I ------")
# ------------------------- 

print(one)
vuelta = one[::-1]
print(vuelta)

# ------------------------- 
print("------ Contar Nucleotidos ------")
# ------------------------- 

two = Seq("ATGGCGAGGAACCCA")

num_A = two.count("A")
print(num_A)

num_T = two.count("T")
print(num_T)

# ------------------------- 
print("------ SEqUtils ------")
# ------------------------- 

from Bio.SeqUtils import gc_fraction
from Bio.SeqUtils import nt_search

tres = Seq("CCGCCGTAAATGGCGAGGAACCCAAA")
subtres = Seq("CC")

print(gc_fraction(two))
print(gc_fraction(tres)) 
print("%0.2f" % gc_fraction(tres))

lista = nt_search(str(tres),str(subtres))
print(lista)
````

# Transcripción

````python
four_cod = Seq("CTATCGGGCACCCTTGCGGCCCATTACAATGGCCAT")

trn_four = four_cod.transcribe()
print(trn_four)

# ------------------------- 
print("------ Hebras ------")
# ------------------------- 

print("La hebra codificante: es la que se replica (5'-> 3') \n" \
    "La hebra molde: es la que se trancribe y traduce (3' -> 5')\n" \
    "Lo que se hace es DAR la vuelta a la molde y transcribir para direccion correcta\n\n")

print(f"Hebra codificante:          {four_cod}") # 5' ->3'
print(f"Hebra molde:                {four_cod[::-1]}") # 3' ->5'
print(f"Transcrito ARN de molde:    {four_cod.reverse_complement().transcribe()}") # 5'->3' + transcripcion
````

# Traducción

````python
one_arn = Seq("AUGGCCAUUGUAAUGGGCCGCAAGGGUGCCCGAUAG")
print(one_arn)

trl_one = one_arn.translate()
print(trl_one)

# ------------------------- 
print("------ Problemas ------")
# ------------------------- 

print("No multiplo 3 -> Warning\n" \
    "Solución: Recortar o Añadir N")

two_arn = Seq("AUGGC")
trl_two = two_arn.translate
print(trl_two)

# ------------------------- 
print("------ Codones ------")
# ------------------------- 
tres_arn = Seq("ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG")
trl_tres = tres_arn.translate()
print(f"Traducción entera: {trl_tres}")

trl_tres_stop1 = tres_arn.translate(to_stop=True)
print(f"Traducción entera: {trl_tres_stop1}")

trl_tres_codend = tres_arn.translate(stop_symbol="$")
print(f"Traducción entera: {trl_tres_stop1}")

print("\nsecuencia codificante -> optimice traducción (cod inicio metionina, multiplo 3 y unico cod final")
four_arn = Seq("AUGCCGACGUGA")
trl_four_cds = four_arn.translate(cds= True)
print(f"Traducción de SeqCodf: {trl_four_cds}")
````

# Tablas Traducción
Para realizar la traducción empleando una determinada tabla, será necesario incluir el argumento `table` en la llamada al método `translate()` de la siguiente forma (ejemplos):
   - id
    
    ```Python
    Seq.translate(table=1)
    Seq.translate(table=11)
    Seq.translate(table=5)
    ```
   - nombre
    
    ```Python
    Seq.translate(table="Standard") 
    Seq.translate(table="Bacterial") 
    Seq.translate(table="Invertebrate Mitochondrial")
    ```

# MutableSeq

````python
from Bio.Seq import MutableSeq

yi = Seq("GGGATTGGTGCAG")
er = MutableSeq("GGGATTGGTGCAG")

# ------------------------- 
print("------ Cambio 1 letra ------")
# ------------------------- 


print(yi[3])
print(er[3])

#yi[3] = "T"            # NO OCURRE PORQUE ES INMUTABLE
print(yi)
er[3] = "T"
print(er)

# ------------------------- 
print("------ Quitar 1º aparicion letra ------")
# ------------------------- 
print(er)
er.remove("G")
print(er)

# ------------------------- 
print("------ Metodos aplicación directa ------")
# ------------------------- 
print("Estos cambios a veces se ven reflejados o no, está fuera de nuestro control." \
"Es un diseño de biopython.\n")
san = MutableSeq("GGGATTGGTGCAG")
print(san)

san.reverse()
print(san)

san[::-1]
print(san)

san.reverse_complement()
print(san)

print(san.reverse_complement())
````
