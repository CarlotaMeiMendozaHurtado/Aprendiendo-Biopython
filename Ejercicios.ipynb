```python
from Bio.Seq import Seq

yi = Seq("ATGGCCATTGTAAATGGGCCC")
print(yi)

#-------------------------------------
print("---------- Solución ----------")
print(yi.count("AA") , yi.count("CC") , yi.count("CCC"))

print(yi.count_overlap("AA") , yi.count_overlap("CC") , yi.count_overlap("CCC"))

```

```python
from Bio.SeqUtils import gc_fraction
san = Seq("GCATGGCCATTGTAATGGGC") 

posicion_ATG = san.find("ATG")
print(posicion_ATG)

despues_ATG = posicion_ATG+3
print(despues_ATG)

print(f"{gc_fraction(san[despues_ATG::])*100:.2f}")
```

```python
from Bio import SeqIO
from Bio.SeqIO import SeqRecord
from Bio.Seq import Seq

ejer_10 = SeqIO.parse("data1C/ejemplo_registros_duplicados.fastq" , "fastq")

for e in ejer_10:
    print(f"{e.letter_annotations}")

    # extraer los elementos, es una lista
    calidades = e.letter_annotations["phred_quality"]

    # calcula
    media = sum(calidades)/len(calidades)
    print(f"Media: {media:.2f} \n--------")  # tambiéne stá round(media,2)
```
