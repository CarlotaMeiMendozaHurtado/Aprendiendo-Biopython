### 4.3 PAM vs BLOSUM

- PAM: 
    - Basada en registros de secuencias muy similares.
    - Con n alto (PAMn): para analizar secuencias divergentes.
    - Con n bajo (PAMn): para analizar secuencias similares.
    - Ejemplo común: PAM250
- BLOSUM: 
    - Basada en observaciones de alineamientos entre regiones conservadas de familias de proteinas
    - Con n alto (PAMn): para analizar secuencias similares.
    - Con n bajo (PAMn): para analizar secuencias divergentes.
    - Ejemplo común: BLOSUM62

## 6. Alg. Prog. Dim

### 6.1 Alg Needleman-Wunsch - ali. Global
1. Creamos matrices N y T
    1. Rellenar N: siempre 2 gaps iniciales -> 0
    2. Recorrer la matriz con T
2. abajo derecha a arriba izquierda
3. Obtienes
    1. Alineamiento
    2. Puntuación

### 6.2 Alg Smith-Waterman - ali. Local
1. Creamos matrices N y T
    1. Rellenar N: siempre 2 gaps iniciales y TODO 0
    2. Recorrer la matriz con T
2. abajo derecha a arriba izquierda
3. Obtienes
    1. Alineamiento
    2. Puntuación
