```python
from Bio import Align
from Bio.Align import substitution_matrices
import numpy as np

#1. crear alineador
alineador_edit = Align.PairwiseAligner()
#2. modo
alineador_edit.mode = "global"
#3. penalizacion
alineador_edit.open_gap_score = -14
alineador_edit.extend_gap_score = -14

#4. crear y configurar matriz subtitucion
alfabeto = "ATCG"
matriz_visual = np.array([
    [5,-2,-3,-11],  
    [-2,2,0,-13],
    [-3,0,12,-5],
    [-11,-13,-5,20]
])
matriz_lista = substitution_matrices.Array(alfabeto, dims=2, data=matriz_visual)
alineador_edit.substitution_matrix = matriz_lista

print(alineador_edit)
print(matriz_lista)
print(alineador_edit.algorithm)
```
