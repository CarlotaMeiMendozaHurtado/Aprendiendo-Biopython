# Alineamiento biopython
Alineamientos por pares (g y l) usando porgramacion dinámica

PairwiseAligner() tiene:
- `match_score`
- `mismatch_score`
- `substitution_matrix`
- `target_***_gap_score`: valores de las diferentes puntuaciones de penalización por gap para la secuencia "target" (por defecto todas =0)
- `query_***_gap_score`: valores de las diferentes puntuaciones de penalización por gap para la secuencia "query" (por defecto todas =0)
- `gap_score`
- `mode`: global o local

# Metodos
Entrada Seq, MutableSeq o SeqRecord
debe haber al menos 2, target y queary

- `score()`: puntuacion
- `aling()`: generador de alineamientos, puede con `len()` y `for`

```python
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio.Seq import MutableSeq
from Bio import Align
from Bio.Align import PairwiseAligner

alineador=Align.PairwiseAligner()
print(alineador.algorithm) # te dice tipo de alineador nw global | sw local

dna_target = Seq("AATG")
target=SeqRecord(dna_target, id="Obj001")

dna_query = Seq("ATT")
query=SeqRecord(dna_query, id="Con001")

alimento = alineador.align(target,query)
print(alimento)
for e in alimento:
    print(e)
```

```python
from Bio import Align
from Bio.Seq import Seq
secuencia_target = Seq('AAAAAAAACCCCCCCCGGGGGGGGTTTTTTTTAAAAAAAACCCCCCCCGGGGGGGGTTTTTTTT')
secuencia_query =  Seq('AAAAAAACCCTCCCCCGGCCGGGTTTAGTTTAAAAAAAACCCCCCCCGGGGGGGGTTTTTTTT')
alineador=Align.PairwiseAligner(mismatch_score=-1,gap_score=-1)
alineamientos=alineador.align(secuencia_target,secuencia_query)
alineamiento=alineamientos[0]
print(alineamiento)
matriz_sus=alineamiento.substitutions
print(matriz_sus)

cuentas=alineamiento.counts()
print(f"El alineamiento tiene {cuentas.gaps} huecos, {cuentas.identities} identidades, {cuentas.mismatches} mismatches")
```
