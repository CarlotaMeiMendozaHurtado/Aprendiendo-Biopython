# SeqRecord

## Atributos

Veamos los diferentes atributos con los que vamos a poder trabajar en esta clase de registros:

- `.seq`. Objeto de tipo `Seq`.  Contiene la secuencia como tal.
- `.id`. Objeto de tipo *string*. ID primario que permite identificar la secuencia. En la gran mayoría de los casos se trata del numero de acceso (`ACCESSION`)
- `.name`. Objeto de tipo *string*.  Contiene el nombre común para la secuencia. Podemos decir que se trata del incluye el nombre del locus o posición en el genoma.
- `.description`. Objeto de tipo *string*.  Se encuentra una breve descripción para la secuencia.
- `.letter_annotations`. Objeto de tipo diccionario.  Contiene anotaciones por cada una de las letras y emplea un diccionario restringido de información adicional sobre las letras de la secuencia. Las claves son el nombre de la información, y la información está contenida en el valor como una colección de Python (es decir, una lista, tupla o cadena) con la misma longitud que la propia secuencia. Esto se utiliza a menudo para incluir valores de calidad o información sobre estructuras secundarias (por ejemplo, en archivos de alineación).
- `.annotations`. Objeto de tipo diccionario.  Contiene información adicional de la secuencia. Las claves contienen el tipo de información y el valor tiene la información como tal.
- `.features`. Objeto de tipo lista de `SeqFeature`.  Contiene información sobre la secuencia de una forma estructurada. Veremos un poco mas adelante los campos de `SeqFeature`.


# Carga manual

```python
one = Seq("ATGGCCATTGTAATGGGC")
reg_one = SeqRecord(one)

reg_one.id = "HXL47"
reg_one.description = "Prueba de esta seq"
reg_one.letter_annotations["valor_calidad"] = [25, 40, 60, 54,12, 14,35,89,34,55,60,44,44,32,78,60,25,31]
reg_one.annotations["comentario"] = "Pruebitas"


print(reg_one.seq)
print(reg_one.id)
print(reg_one.description)
print("------------Anotacion, diccionario,nombre------------")
print(reg_one.letter_annotations)
print(reg_one.letter_annotations.keys())
print(reg_one.annotations)
print(reg_one.annotations.values())
print(reg_one.annotations.keys())
```

# Archivos Comprimidos
- `.gz`
- `.bz2`
- `.bgz`

```python
print("------gzip1------")
import gzip
total = 0
with gzip.open("data1C/ejemplo_registros.fastq.gz" , "rt") as archivo_1a:
    for e in SeqIO.parse(archivo_1a,"fastq"):
        total += len(e)
        print(total)



print("------gzip2------")
import gzip
with gzip.open("data1C/ejemplo_registros.fastq.gz" , "rt") as archivo_1b:
    print(sum(len(e) for e in SeqIO.parse(archivo_1b, "fastq")))
# print(sum(xxxxx))



print("------bz2------")
import bz2
with bz2.open("data1C/ejemplo_registros.fastq.bz2" , "rt") as archivo_2:
    print(sum(len(e) for e in SeqIO.parse(archivo_2,"fastq")))



print("------bgz------")  
from Bio import bgzf
with bgzf.open("data1C/ejemplo_registros.fastq.bgz" , "rt") as archivo_3:
    print(sum(len(e) for e in SeqIO.parse(archivo_3,"fastq")))
```
